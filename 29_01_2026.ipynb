## **Nested Conditional Statement**



```
4ï¸âƒ£ Nested Conditional Statement
ðŸ”¹ Definition

When an if statement is written inside another if or else block.

ðŸ”¹ Syntax
if condition1:
    if condition2:
        statement(s)
    else:
        statement(s)
else:
    statement(s)

ðŸ”¹ When to use

When a condition depends on another condition
For hierarchical decision-making
```



````python
age = 20
id_card = True

if age >= 18:
    if id_card:
        print("Entry allowed")
    else:
        print("ID card required")
else:
    print("Underage")
````
