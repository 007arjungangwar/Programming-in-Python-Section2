# **Problem: Pangram Check**
```
Note: A pangram contains all 26 letters of English alphabet.

````python
sentence="The quick brown fox jumps over the lazy dog"
temp=""
for ch in sentence:
    if "a"<=ch<="z" or "A"<=ch<="z":
        if ch not in temp:
            temp+=ch.lower() # convert either in upper or lower case

print(temp)

if len(temp)==26:
    print("Pangram")
else:
    print("not pangram")
````

## Parameters Vs Arguments

- Vars in `()` during func definition..
- Defined in func declaration.
  
```python

### Arguments:
- Values passed at func call.
- Inputs during function invocation.
  
```python

```

*2. Positional Argument*

## **Create a function which can multiply any number of input and return you the product of all those numbers.**

````python
# Useful when the number of arguments is unknown.
def multiply(*number):
    product = 1
    for i in number:
        product *= i
    print(product)
````

````python
multiply(1, 2)
````

````python
multiply(1, 2, 3)
````

## `*args` and `**kwargs`

`*args`: Variable-length positional arguments. (Basically create **tuple**)
```python
def func(*args)
```
`**kwargs`: Variable-length keyword arguments. (Basically create **dictionary** based on the key & value)
```python
def func(**kwargs)
```

## **Create a function which can take any number of input and return you the list of cube of all those numbers.**

````python
def Cube(*arg):
    temp=[]
    for i in arg:
        temp.append(i**3)

    return temp
````

## **Create Function which can Take any number of Input as a string and return, All those string start with the character "A" If not found then return not found**

**Let's solve first without the function and after that we will convert into the function**

````python
# list1=["Arjun","Meena","Arun","Reena","Pankaj"]
list1=["Meena","Reena","Pankaj"]
list2=[]
for i in list1:
  if i.startswith("A"):
    list2.append(i)
if len(list2)!=0:
  print(list2)
else:
  print("Not found")
````

````python
def Word_start_with_special_ch(*args): #name args not Mandatory.
  list1=args # basically this is a tuple
  list2=[]
  for i in list1:
    if i.startswith("A"):
      list2.append(i)
  if len(list2)!=0:
    return list2
  else:
    return "Not found"
````
