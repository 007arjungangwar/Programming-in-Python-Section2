## Lists vs Tuples

### Syntax:
  - Lists: `[ ]`  
  - Tuples: `( )`

### Mutability:
  - Lists: Mutable  
  - Tuples: Immutable

### Speed:
  - Lists: Slower (mutable)  
  - Tuples: Faster (immutable)

### Memory:
  - Lists: Higher  
  - Tuples: Lower

### Functionality:
  - Both: Indexing, slicing  
  - Lists: More methods

### Error-Prone:
  - Lists: Modifiable  
  - Tuples: Safer

### Use Case:
  - Lists: Dynamic  
  - Tuples: Static

````python
import time
L = list(range(100000000))
T = tuple(range(100000000))

# List timing
start = time.time()
for i in L:
  i*5
end=time.time()
print('List time', end-start)
````

````python
# Tuple timing
start = time.time()
for i in T:
  i*5
end=time.time()
print('Tuple time', end-start)
````

````python
import sys
L = list(range(1000))
T = tuple(range(1000))
print('List size', sys.getsizeof(L))
print('Tuple size', sys.getsizeof(T))
````

## **Zip**

````python
l1=list(range(1,6))
l2=list(range(5,11))
l3=list(range(11,16))
print(l1)
print(l2)
print(l3)
zip(l1,l2)
for i in zip(l1,l2):
    print(i)
````

````python
l1=list(range(1,6))
l2=list(range(5,11))
l3=list(range(11,16))
print(l1)
print(l2)
print(l3)
zip(l1,l2,l3)
for i in zip(l1,l2,l3):
    print(i)
````

````python
l1=tuple(range(1,6))
l2=tuple(range(5,11))
l3=tuple(range(11,16))
print(l1)
print(l2)
print(l3)
zip(l1,l2,l3)
for i in zip(l1,l2,l3):
    print(i)
````

````python
l1=tuple(range(1,6))
l2=tuple(range(5,11))
l3=tuple(range(11,16))
print(l1)
print(l2)
print(l3)
for i,j,k in zip(l1,l2,l3):
    print(i,j,k)
````
