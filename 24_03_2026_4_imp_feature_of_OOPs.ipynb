## **Abstraction - hiding information by function(method or class)**

````python
# Abstraction - hiding information by function(method or class)

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.percentage=percentage

    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {round((self.percentage+5)/1.1,2)}%"

    # hiding info how pecentage is calculating

s2=Student("Roshni",12,92)
s2.student_info()
````

**Here we are just showing the percentage to the user Not telling how we are calculating So actually by the method we are hiding the process.**

## **Encapsulation: - restrict access of some attribute or method**

````python
# Encapsulation: - restrict access of some attribute or method

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.__percentage=percentage

    def get_percentage(self):
        return self.__percentage

    def set_percentage(self,temp): # Instead of temp you can pass any of the name
        if 0<=temp<=100:
            self.__percentage=temp
            return self.__percentage
        else:
            return ("Not Allowed")


    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {self.get_percentage()}%"
````

````python
s1=Student("Meera",11,52)
print(s1.name,s1.grade,s1.get_percentage())
````

## **Inheritance - inherit property of class by super**

````python
# Inheritance - inherit property of class by super

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.percentage=percentage

    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {round((self.percentage+5)/1.1,2)}%"


class GraduateStudent(Student):
    def __init__(self, name, grade, percentage, subject, marks, section, stream):
        super().__init__(name,grade,percentage)
        self.marks=marks
        self.section=section
        self.stream=stream
        self.subject=subject
        super().student_info()
````

````python
class Line:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c

    def __str__(self):
        return f"{self.a}x+{self.b}y+{self.c}=0"

    def prallel_line(self,other):
        lhl=self.a/self.b
        rhl=other.a/other.b
        if lhl==rhl:
            return f"Lines {self.a}x+{self.b}y+{self.c}=0 and {other.a}x+{other.b}y+{other.c}=0 are parallel to each other."
        else:
            return f"Lines {self.a}x+{self.b}y+{self.c}=0 and {other.a}x+{other.b}y+{other.c}=0 are not parallel."
````

````python
l1=Line(1,2,3)
print(l1)

l2=Line(2,2,1)
print(l2)
````

````python
print(l1.prallel_line(Line(2,2,1)))
print(l1.prallel_line(l2))
print(l2.prallel_line(l1))
````

````python
class Animal:
    def __init__(self,name):
        self.name=name
                            # Parent class
    def eat(self):
        return ("Eating")

    def sleep(self):
        return ("Parent Sleeping")

class Dog(Animal):
    def __init__(self,color,name):
        super().__init__(name)
        self.color=color             # Child class

    def bark(self):
        return ("Barking")
````

## **Type 1: Method Overriding**

```
Child class modifies parent class method.
```



````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")
````

````python
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)    # Call parent constructor
        self.student_id = student_id

    def display(self):
        super().display()               # Call parent method
        print(f"Student ID: {self.student_id}")

student = Student("Arjun", 20, "S123")
student.display()
````

````python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):           # Overriding
        print("Dog barks")

class Cat(Animal):
    def sound(self):           # Overriding
        print("Cat meows")

animals = [Animal(), Dog(), Cat()]
for animal in animals:
    animal.sound()
````

````python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Woof! Woof!")

class Cat(Animal):
    def sound(self):
        print("Meow! Meow!")

class Cow(Animal):
    def sound(self):
        print("Moo! Moo!")

animals = [Animal(), Dog(), Cat(),Cow()]
for animal in animals:
    animal.sound()
````

### **Type 2: Method Overloading (Simulated in Python)**

```
Same method name, different parameters.
```



````python
class Multiply:

    def multiply(self,a,b):
        return a*b
    def multiply(self,a,b,c):
        return a*b*c
````

````python
m1=Multiply()
m1.multiply(1,2)
````

````python
class Multiply:

    def multiply(self,a,b,c=1):
        return a*b*c
````

````python
m1=Multiply()
m1.multiply(1,2)
````

````python
m1=Multiply()
m1.multiply(1,2,7)
````

````python
class Multiply:
    def multiply(self,*args):
        mult=1
        for i in args:
            mult*=i

        return mult
````

````python
m2=Multiply()
m2.multiply(1,2,3,4)
````

````python
class Calculator:
    # Using default arguments
    def add(self, a, b=0, c=0):
        return a + b + c

    # Using *args (best way)
    def multiply(self, *numbers):
        result = 1
        for num in numbers:
            result *= num
        return result
````

````python
calc = Calculator()
print(calc.add(5))          # Output: 5
print(calc.add(5, 10))      # Output: 15
print(calc.add(5, 10, 15))  # Output: 30
````

````python
print(calc.multiply(2, 3))      # Output: 6
print(calc.multiply(2, 3, 4))   # Output: 24
````

### **Type 3: Operator Overloading**

```
Changing behavior of operators for user-defined classes.
or
Changing behavior of operators (+, −, ∗, etc.)
```



````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Overloading + operator
    def __add__(self, other):
        return self.marks + other.marks

s1 = Student("Arjun", 85)
s2 = Student("Rahul", 70)

print(s1 + s2)        # Output: 155
````

````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Overloading + operator
    def __add__(self, other):
        return self.marks + other.marks

    # Overloading > operator
    def __gt__(self, other):
        return self.marks > other.marks

s1 = Student("Arjun", 85)
s2 = Student("Rahul", 70)

print(s1 + s2)        # Output: 155
print(s1 > s2)        # Output: True
````
