## **Abstraction - hiding information by function(method or class)**

````python
# Abstraction - hiding information by function(method or class)

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.percentage=percentage

    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {round((self.percentage+5)/1.1,2)}%"

    # hiding info how pecentage is calculating

s2=Student("Roshni",12,92)
s2.student_info()
````

## **Encapsulation: - restrict access of some attribute or method**

````python
# Encapsulation: - restrict access of some attribute or method

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.__percentage=percentage

    def get_percentage(self):
        return self.__percentage

    def set_percentage(self,temp): # Instead of temp you can pass any of the name
        if 0<=temp<=100:
            self.__percentage=temp
            return self.__percentage
        else:
            return ("Not Allowed")


    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {self.get_percentage()}%"
````

````python
s1=Student("Meera",11,52)
print(s1.name,s1.grade,s1.get_percentage())
````

## **Inheritance - inherit property of class by super**

````python
# Inheritance - inherit property of class by super

class Student:
    def __init__(self,name, grade, percentage):
        self.name=name
        self.grade=grade
        self.percentage=percentage

    def student_info(self):
        return f"{self.name} is in {self.grade} grade and percentage is {round((self.percentage+5)/1.1,2)}%"


class GraduateStudent(Student):
    def __init__(self, name, grade, percentage, subject, marks, section, stream):
        super().__init__(name,grade,percentage)
        self.marks=marks
        self.section=section
        self.stream=stream
        self.subject=subject
        super().student_info()
````

````python
s1=GraduateStudent("Nitish",12,55,["Maths", "English", "Computer"],[95,92,95] ,"section2","PCM")
s2=GraduateStudent("Advika",11,98,["Maths", "English", "Bio"],[90,97,82] ,"section2","PCB")
````

````python
print(s1.subject)
print(s1.marks)
print(s1.grade)
print(s1.student_info())
````

````python
print(s2.subject)
print(s2.marks)
print(s2.grade)
print(s2.student_info())
````

## **Polymorphism : same method name but working is different**

```
Check Parallel Lines
Create a method to:
1)Check if two lines are parallel

Condition:
a1/b1 = a2/b2
```



````python
class Line:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c

    def __str__(self):
        return f"{self.a}x+{self.b}y+{self.c}=0"

    def prallel_line(self,other):
        lhl=self.a/self.b
        rhl=other.a/other.b
        if lhl==rhl:
            return f"Lines {self.a}x+{self.b}y+{self.c}=0 and {other.a}x+{other.b}y+{other.c}=0 are parallel to each other."
        else:
            return f"Lines {self.a}x+{self.b}y+{self.c}=0 and {other.a}x+{other.b}y+{other.c}=0 are not parallel."
````

````python
l1=Line(1,2,3)
print(l1)

l2=Line(2,2,1)
print(l2)
````

````python
print(l1.prallel_line(Line(2,2,1)))
print(l1.prallel_line(l2))
print(l2.prallel_line(l1))
````

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")
````

````python
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)    # Call parent constructor
        self.student_id = student_id

    def display(self):
        super().display()               # Call parent method
        print(f"Student ID: {self.student_id}")

student = Student("Arjun", 20, "S123")
student.display()
````
