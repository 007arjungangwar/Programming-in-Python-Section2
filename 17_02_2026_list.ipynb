# **String indexing and slicing**

````python
c = "Education is one of the key of Success"

print(c[0])        # prints the first character of the string
print(c[-1])       # prints the last character of the string

print(c[0:5])      # prints characters from index 0 to 4
print(c[2:])       # prints the string starting from index 2 till the end
print(c[:4])       # prints characters from the start up to index 3
print(c[:])        # prints the complete string

print(c[2:6:2])    # prints characters from index 2 to 5, skipping every 2nd character
print(c[0:8:3])    # prints characters from index 0 to 7, skipping every 3rd character

print(c[0:6:-1])   # prints nothing because step is negative but start < stop
print(c[-5:-1:2])  # prints characters from index -5 to -2, skipping every 2nd character

print(c[-1:-5:-1]) # prints characters from last index to index -4 in reverse order
print(c[::-1])     # prints the entire string in reverse order
````

## Arrays vs. Lists in Memory:

- **Arrays:**
  - `int arr[50];` ---> 50 contiguous memory blocks.
  - Elements stored in binary form at consecutive addresses.

- **Lists:**
  - `list_example = [1, 2, 3]` ---> Elements at different locations.
  - Stores references/pointers to elements, not the values.

````python
# proof --->
L = [1, 2, 3]
print(id(L))        # Prints the memory address of the list object itself.
print(id(L[0]))     # Prints the memory address of the integer object 1.
print(id(L[1]))     # Prints the memory address of the integer object 2.
print(id(L[2]))     # Prints the memory address of the integer object 3.
print(id(1))        # Prints the memory address of the integer object 1.
print(id(2))        # Prints the memory address of the integer object 2.
print(id(3))        # Prints the memory address of the integer object 3.
````



```
# Observe that id(L[0]) is the same as id(1), id(L[1]) is the same as id(2), and so on.
# This demonstrates that lists store references/pointers to elements, not the values directly.
# The elements themselves (the integers 1, 2, 3) are distinct objects in memory,
# and the list 'L' contains references to these integer objects at different locations.
```

````python
# Check if list is ascending
def is_ascending(L):
    for i in range(len(L) - 1):
        if L[i] > L[i + 1]:
            return False
    return True

# Test
L1 = [1, 2, 3, 4, 5]
L2 = [1, 3, 2, 4, 5]

print("L1 is in ascending order:", is_ascending(L1))
print("L2 is in ascending order:", is_ascending(L2))
````

## **Transpose a matrix**

````python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transpose)
````

````python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Determine dimensions
rows = len(matrix)
cols = len(matrix[0])

# Initialize an empty list for the transpose
transpose = []

# Iterate through columns of the original matrix
for i in range(cols):
    new_row = []
    # Iterate through rows of the original matrix
    for j in range(rows):
        # Append the element at (j, i) to the new row
        new_row.append(matrix[j][i])
    # Add the constructed row to the transpose matrix
    transpose.append(new_row)

print(transpose)
````

## **55 Identity Matrix**

````python
matrix = [[1 if i == j else 0 for i in range(5)] for j in range(5)]
print(matrix)
````

````python
# Initialize the outer list
matrix = []

# Outer loop (j) represents the rows
for j in range(5):
    row = []
    # Inner loop (i) represents the columns
    for i in range(5):
        # If the row index matches the column index, use 1
        if i == j:
            row.append(1)
        # Otherwise, use 0
        else:
            row.append(0)
    # Add the completed row to the matrix
    matrix.append(row)

print(matrix)
````
