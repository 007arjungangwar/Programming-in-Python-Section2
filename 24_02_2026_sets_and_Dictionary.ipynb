## Frozen set

- Immutable set.
- Created via `frozenset()`.
- No modifications post-creation.
- Ideal for hashable items: **strings**, **numbers**.

````python
# create frozenset
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])

print(fs1,type(fs1))
print(fs2,type(fs2))
````

````python
# create frozenset
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])
fs1 | fs2
````

````python
# 5. Frozen sets (immutable sets)
frozen = frozenset([1, 2, 3, 4, 5])

print(f"Frozen set: {frozen}")
print(f"Type: {type(frozen)}")
````

````python
# Frozen set operations (read-only)
frozen2 = frozenset([4, 5, 6, 7, 8])
print(f"Union of frozen sets: {frozen | frozen2}")
````

## **Count the character in the word or sentence**

````python
# By using nested for loop with if condition
word = "sliceeliss"

for ch in word:
    count = 0
    for i in word:
        if i == ch:
            count += 1
    print(f"Word: {ch} and their count: {count}")
````

````python
# Removing duplicate by using list
word = "sliceeliss"
seen = []

for ch in word:
    if ch not in seen:        # ensure it runs only first time
        count = 0
        for i in word:
            if i == ch:
                count += 1
        print(f"Word: {ch} and their count: {count}")
        seen.append(ch)       # store processed character
````

````python
# Removing duplicate by using set
word = "sliceeliss"
seen = set()

for ch in word:
    if ch not in seen:
        count = 0
        for i in word:
            if i == ch:
                count += 1
        print(f"Word: {ch} and their count: {count}")
        seen.add(ch)
````

````python
# Removing duplicate by using dict
word = "sliceeliss"
freq = {}

for ch in word:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1

for ch, count in freq.items():
    print(f"Word: {ch} and their count: {count}")
````

## Dictionary Comprehension

```python
{key : value for var in iterable}
```

Creates dicts from iterables.

**Example:**

`squares_dict = {x: x**2 for x in range(5)}`

Maps numbers to squares.

Supports conditions & expressions.

````python
D = {"Name":"Arjun", "Gender":"Male", "Age":28}
print(D)
print(D.items())
````

````python
D1 = {key:value for key, value in D.items() if len(key)>3}
print(D1)
````

````python
L = [1, 2, 3, 4, 5, 6, 7]
D2 = {item : item**2 for item in L} # first becomes keys : second becomes values
D2
````

````python
D2 = {item : item**2 for item in L if item%2 == 0} # first becomes keys : second becomes values
D2
````

````python
# Print 1st 10 nums & squares
{i:i**2 for i in range(1, 11)} # first becomes keys : second becomes values
````

````python
distances = {'delhi':1000, 'mumbai':2000, 'bangalore':3000}
print(distances.items())
````

````python
# using existing dict
distances = {'delhi':1000, 'mumbai':2000, 'bangalore':3000}
{key:value*0.6 for (key, value) in distances.items()}
````
