**Implicit Conversion**
- Python automatically converts between compatible types:

```python
# Integer to Float
result = 5 + 3.2
print ( " 5 + 3.2 = " , result ) # 8.2
print ( " Type : " , type ( result ) ) # < class ’ float ’>
# Integer to Complex
result2 = 4 + (2 + 3j)
print ( " 4 + (2+3j ) = " , result2 ) # (6+3 j )
print ( " Type : " , type ( result2 ) ) # < class ’ complex ’>
```

**Explicit Conversion**
- You can manually convert between types:

```python
# String to Integer
num_str = " 123 "
num_int = int ( num_str )
print ( num_int , type ( num_int ) ) # 123 < class ’ int ’>
# Integer to String
age = 25
age_str = str ( age )
print ( age_str , type ( age_str ) ) # ’25 ’ < class ’ str ’>
# Float to Integer
price = 9.99
price_int = int ( price ) # Note : This truncates , not rounds
print ( price_int ) # 9
# List to Set
numbers = [1 , 2 , 2 , 3 , 3 , 3]
unique = set ( numbers )
print ( unique ,type ( unique ) ) # {1 , 2 , 3}
```

- **Arithmetic**: `+`, `-`, `*`, `/`, `%`, `**`, `//`

```python
x = 10
y = 3
print ( " Addition : " , x + y ) # 13
print ( " Subtraction : " , x - y ) # 7
print ( " Multiplication : " , x * y ) # 30
print ( " Division : " , x / y ) # 3.333...
print ( " Floor Division : " , x // y ) # 3
print ( " Modulus : " , x % y ) # 1
print ( " Exponent : " , x ** y ) # 1000
```

- **Comparison**: `==`, `!=`, `>`, `<`, `>=`, `<=`


```python
a = 5
b = 3
print ( " a > b : " , a > b ) # True
print ( " a < b : " , a < b ) # False
print ( " a == b : " , a == b ) # False
print ( " a != b : " , a != b ) # True
print ( " a >= b : " , a >= b ) # True
print ( " a <= b : " , a <= b ) # False
```

- **Logical**: `and`, `or`, `not`


```python
x = True
y = False
print ( " x and y : " , x and y ) # False
print ( " x or y : " , x or y ) # True
print ( " not x : " , not x ) # False
print ( " not y : " , not y ) # True
```

- **Assignment**: `=`, `+=`, `-=`, `*=`, `/=`, `%=`, `**=`, `//=`


```python
x = 5
print ( " Initial x : " , x )
x += 3 # Same as x = x + 3
print ( " After x += 3: " , x )
x -= 2 # Same as x = x - 2
print ( " After x -= 2: " , x )
x *= 4 # Same as x = x * 4
print ( " After x *= 4: " , x )
x /= 2 # Same as x = x / 2
print ( " After x /= 2: " , x )
```
