# Slicing



```
üîπ 3. String Slicing

Slicing extracts a part of the string.
Syntax
string[start : end : step]

```



````python
word="Hello World"
print(len(word))
word[:15]
````

````python
word="Hello World"

print(word[0:6]) # Slices from index 0 up to (but not including) index 6, resulting in 'Hello '.
print(word[2:]) # Slices from index 2 to the end of the string, resulting in 'llo World'.
print(word[:6]) # Slices from the beginning up to (but not including) index 6, resulting in 'Hello '.
print(word[2:6]) # Slices from index 2 up to (but not including) index 6, resulting in 'llo '.
print(word[2:15:2]) # Slices from index 2 to the end of the string with a step of 2, resulting in 'loWrd'.
````

- Negative slicing

````python
word="Hello World"

print(word[-7:-1]) # Slices from the 7th character from the end up to (but not including) the 1st character from the end. This prints 'o Worl'.
print(word[-9:-2:2]) # Slices from the 9th character from the end up to (but not including) the 2nd character from the end, with a step of 2. This prints 'loWr'.
print(word[::-1]) # Reverses the entire string. This prints 'dlroW olleH'.
print(word[-9:-1:-2]) # This slice has a negative step but the start index (-9) is less than the end index (-1). Since the step is negative, it expects the start index to be greater than the end index. This results in an empty string.
print(word[2:9:-1]) # This slice has a negative step but the start index (2) is less than the end index (9). Since the step is negative, it expects the start index to be greater than the end index. This also results in an empty string.
print(word[-1:-9:-2])
print(word[9:2:-1])
````

## **10. Check if Two Strings are Anagrams**

```
(Anagram ‚Üí same characters, different order
Ex: "listen" and "silent")

üß† Thinking Process

If lengths not equal ‚Üí not anagram
Count characters manually
Compare frequency
```



**by While loop**

````python
# s1 = input("Enter first string: ")
# s2 = input("Enter second string: ")
s1="listen"
s2="silent"

if len(s1) != len(s2):
    print("Not Anagram")
else:
    i = 0
    is_anagram = True

    while i < len(s1):
        if s1.count(s1[i]) != s2.count(s1[i]):
            is_anagram = False
            break
        i += 1

    if is_anagram:
        print("Anagram")
    else:
        print("Not Anagram")
````

**by nested for loop**

````python
s1="listen"
s2="silenY"

if len(s1)!=len(s2):
  print("Not the Angram")
else:
  is_Angram=False
  for ch in s1:
    for ch in s2:
      is_Angram=True

  if is_Angram:
    print("Angram")
  else:
    print("Not Angram")
````

**by single for loop with conditional statement.**
