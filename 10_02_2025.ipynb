#### **2. Relational Operations (lexicographical comparison)**

````python
# 2. Relational Operations (lexicographical comparison)
print("Apple" < "Banana") # True (A comes before B)
print("apple" < "Banana") # False (lowercase > uppercase in ASCII)
print("Python" == "python") # False (case-sensitive)
print("Hello" != "World")
````

````python
# 3. Logical Operations
print("Hello" and "World") # World
print("" or "Default") # Default
print(not "Hello") # False
print(not "")
````

#### **4. Membership Operations**

````python
# 4. Membership Operations
quote = "To be or not to be"
print("be in quote:", "be" in quote) # True
print("python in quote:", "python" in quote) # False
print("BE in quote:", "BE" in quote) # False (case-sensitive)
print("to not in quote:", "to" not in quote) # True (case-sensitive)
````

## **6. Important String Methods**

### **1.  Case conversion (upper, lower, title, capitalize, swapcase)**

````python
text = " python programming is FUN! "
# 1. Case conversion
print("Original:", text)
print("upper():", text.upper()) # Converts all characters in the string to uppercase
print("lower():", text.lower()) # Converts all characters in the string to lowercase
print("title():", text.title()) # Converts the first character of each word to uppercase and the rest to lowercase
print("capitalize():", text.capitalize()) # Capitalizes the first character of the string and converts the rest to lowercase
print("swapcase():", text.swapcase()) # Swaps the case of all characters (uppercase becomes lowercase and vice versa)
````

````python
word="     ARjuN GAngwAr  "
print("Original:", word)
print("upper():", word.upper())
print("lower():", word.lower())
print("title():", word.title())
print("capitalize():", word.capitalize())
print("swapcase():", word.swapcase())
````

### **2. Stripping whitespace**

````python
# 2. Stripping whitespace
print("With strip():", text.strip()) # Removes leading and trailing whitespace
print("With lstrip():", text.lstrip()) # Removes leading whitespace
print("With rstrip():", text.rstrip()) # Removes trailing whitespace
````

````python
print("With strip():", word.strip())
print("With lstrip():", word.lstrip())
print("With rstrip():", word.rstrip())
````

### **3. Finding and replacing**

````python
# 3. Finding and replacing
sentence = "The quick brown fox jumps over the lazy dog"
print('count("o"):', sentence.count("o")) # Counts the occurrences of a substring
print('find("fox"):', sentence.find("fox")) # Finds the first occurrence of a substring and returns its starting index, or -1 if not found
print('index("fox"):', sentence.index("fox")) # Similar to find(), but raises a ValueError if the substring is not found
print('replace("fox", "cat"):', sentence.replace("fox", "cat")) # Replaces all occurrences of a specified substring with another
````

### **4. Checking string properties (isalnum, isalpha, isdigit, islower, isupper)**

````python
# 4. Checking string properties
# \n is used here for new line or line change
print("Original:", text)

print("\n'Python123'.isalnum():", "Python123".isalnum()) # Checks if all characters in the string are alphanumeric (letters or numbers)
print("'Python'.isalpha():", "Python".isalpha())     # Checks if all characters in the string are alphabetic (letters)
print("'123'.isdigit():", "123".isdigit())       # Checks if all characters in the string are digits
print("'python'.islower():", "python".islower())     # Checks if all cased characters in the string are lowercase
print("'PYTHON'.isupper():", "PYTHON".isupper())     # Checks if all cased characters in the string are uppercase
````

### **5. Splitting and joining**

````python
# 5. Splitting and joining
data = "apple,banana,cherry,date"
fruits = data.split(",") # Splits the string by the specified delimiter into a list of substrings
print("\nSplit fruits:", fruits) # \n is used here for new line or line change
print("Joined with ‚Äô-‚Äô:", "-".join(fruits)) # Joins elements of an iterable (e.g., a list) with a specified string as a separator
print("Joined with ‚Äô@‚Äô:", "@".join(fruits)) # Joins elements of an iterable (e.g., a list) with a specified string as a separator
````

## **use of .format() method**

````python
# 6. String formatting
name = "Alice"
age = 25


# Using the .format() method
print("format(): {} is {} years old".format(name, age))
print("format(): {name} is {age} years old".format(name="Alice", age=25))
print("format(): {age} is {name} years old".format(name="Alice", age=25))
print("format(): {0} is {1} years old".format("Alice", 25))
print("format(): {1} is {0} years old".format("Alice", 25))


# f-string literal (Python 3.6+), for direct embedding of expressions
print("f-string: {name} is {age} years old")
print(f"Using f-string: {name} is {age} years old")


# Old style formatting using the % operator
print("\nOld style: %s is %d years old" % (name, age))
````

````python
name = "Arjun"
age = 28

print(f"My name is {name} and age is {age}")
print(f"My name is {age} and age is {name}")

print("My name is {} and age is {}".format(name, age))
print(f"My name is {name} and age is {age}".format(name, age))
print(f"My name is {age} and age is {name}".format(name, age))
````

# **3. Palindrome Check (String)**

```
üß† Logic Building

Reverse the string
Compare with original
```



````python
s = input("Enter string: ")

rev = ""
for ch in s:
    rev = ch + rev

if s == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
````

````python
s = input("Enter string: ")

if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
````


# **4. Count Words**
```
üß† Logic Building

Words are separated by spaces
Count transitions from space to character
```



````python
s = input("Enter sentence: ")

count = 0
in_word = False

for ch in s:
    if ch != " " and not in_word:
        count += 1
        in_word = True
    elif ch == " ":
        in_word = False

print("Word count:", count)
````

Traverse string
Check each character is vowel or not

````python
s = input("Enter string: ")
vowels = "aeiouAEIOU"

count = 0
for ch in s:
    if ch in vowels:
        count += 1

print("Vowel count:", count)
````

# **6. Frequency of Characters**

```
üß† Logic Building

Use dictionary
Count occurrences
```



````python
text = "programming is fun"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Text:", text)
print("Character frequencies:")

for key, value in frequency.items():
    print(f"{key} : {value}")
````

## **Program7: Extract Username from Email**

````python
# Extract username from email address

emails = [
    "john.doe@gmail.com",
    "alice.smith@company.co.in",
    "bob@example.org",
    "not-an-email",
    "user@domain"
]

print("Email Username Extraction")
print("=" * 40)

for email in emails:
    if "@" in email:
        at_position = email.index("@")
        username = email[:at_position]
    else:
        username = email

    print(f"{email:30} -> {username}")
````
