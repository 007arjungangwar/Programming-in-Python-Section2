## **Q5**

```
3 digit Armstrong Number
ðŸ’­ How to Think

Get digits
Cube each digit
Compare with original
```



````python
n = 153
temp = n
sum = 0
# sum of each digit Cube equal to the number = Armstrong
while n > 0:
    digit = n % 10
    sum += digit ** 3
    n //= 10

if sum == temp:
    print("Armstrong")
else:
    print("Not Armstrong")
````
