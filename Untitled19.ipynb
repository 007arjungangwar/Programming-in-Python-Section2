```python
class DigitalClock:
    def __init__(self,h,m,s):
        self.h=h
        self.m=m
        self.s=s
    def display(self):
        print(f"{self.h}:{self.m}:{self.s}")
    def tick(self):
        self.s+=1
        if self.s==60:
            self.m+=1
            self.s=00
        if self.m==60:
            self.h+=1
            self.m==00
        if self.h==24:
            self.h=00
            self.s+=1
```

```python
clock1=DigitalClock(23, 59, 59)
clock1.display()
clock1.tick()
clock1.display()
```
