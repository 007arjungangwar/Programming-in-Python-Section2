```python
class SecurityError(Exception):
  def __init__(self,message):
    print(message)
  def logout(self):
    print("logout from all the devices.")
class LoginError(Exception):
  def __init__(self,message):
    print(message)

class Google:
  def __init__(self,name, email, password, device):
    self.name=name
    self.email=email
    self.password=password
    self.device=device

  def login(self,email,password,device):
    if device!=self.device:
      raise SecurityError("Someone trying to login with your account")
    elif email==self.email and password==self.password:
      print(f"Hello {self.name} Welcome back")
    elif email!=self.email and password==self.password:
      raise LoginError("Email is incorrect")
    elif email==self.email and password!=self.password:
      raise LoginError("Password incorrect")
    else:
      raise LoginError("Please check your credentials")

Oppo=Google("Arjun","arjun@gmail.com","123","Android")
try:
  Oppo.login("arjun33@gmail.com","1234","Android")  # Android Windows
except SecurityError as e:
  e.logout()
  print("You are under risk")
except LoginError as p:
  pass
else:
  print(f"Hi {Oppo.name} you are ready to use your device.")
finally:
  print("Database connection closed.")
```
