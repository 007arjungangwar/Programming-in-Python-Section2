```python
# 4 feature of oops
"""
1. Abstraction - hiding unnecessary detail from user by method or Class
2. Encapsulation - Restrict access of certain attribute or methods to protect data and enforce controlled access (Data Protection) private
3. Inheritance - Reuse the property or method from the parent class
4. Polymorphism -  Allow method in different classes to have same name but different behaviour According to object Referred to class.
"""
```

```python
# 1. Abstraction - hiding unnecessary detail from user by method or Class

# Blue print
class Student:
    def __init__(self,name,grade,percentage):
        self.name=name  # property or attribute
        self.grade=grade
        self.percentage=percentage

    def Student_detail(self):
        return f"{self.name} is in class {self.grade} with percentage {self.percentage+2}%"

Ram=Student("Ram",12,89)

Ram.Student_detail()
```

**Here we are just showing the percentage to the user Not telling how we are calculating So actually by the method we are hiding the process.**

```python
# 2. Encapsulation - Restrict access of certain attribute or methods to protect data and enforce controlled access (Data Protection) private

# Blue print
class Student:
    def __init__(self,name,grade,percentage):
        self.name=name  # property or attribute
        self.grade=grade
        self.__percentage=percentage

    def get_percentage(self):
        return self.__percentage

    # Method
    def Student_detail(self):
        return f"{self.name} is in class {self.grade} with percentage {self.get_percentage()+2}%"
    #Object of class student
Ram=Student("Ram",12,89)
Ram.Student_detail()
```

**By the double underscore(__) we are hiding, Which we don't want to show the user, Basically user can not modify those things Directly**

```python
# Inheritance (Reuse the property or method from the parent class)
# Parent class
# Blue print
class Student:
    def __init__(self,name,grade,percentage):
        self.name=name  # property or attribute
        self.grade=grade
        self.percentage=percentage

    # Method
    def student_detail(self):
        return (f"{self.name} is in class {self.grade} and percentage is {self.percentage}%")

# child class
# GraduateStudent is a child class which is inherited the Student class
class GraduateStudent(Student):
    def __init__(self,name, grade, percentage, stream): # i want to add stream
        # parameter from parent class
        super().__init__(name,grade,percentage) #if you dont want you can remove previous attribute
        # new parameter/ attribute use in child class
        self.stream=stream

    def student_detail(self):  # method inherit, we can change the method name
        print(super().student_detail())
        return (f"stream is {self.stream}")


# objects from student class
stud1=Student("Arjun",11,92)
stud2=Student("Mithun",12,96)

print(stud1.student_detail())
print(stud2.student_detail())
```

```python
# objects from GraduateStudent class
g_student1=GraduateStudent("Ram",12,96,"pcm")
g_student2=GraduateStudent("Shyam",11,91,"Bio")
print(g_student1.student_detail())
print(g_student1.stream)
print(g_student1.name)
print(g_student1.percentage)

print(g_student2.student_detail())
print(g_student1.stream)
print(g_student1.name)
print(g_student1.percentage)
```

```python
# polymorphism - Allow method in different classes to have same name but different behaviour According to object Referred to class.

class Student:
    def __init__(self, name, grade, percentage):  # called constructor
        self.name=name # attribute is called Propertiy of class
        self.grade=grade # attribute
        self.percentage=percentage # Double underscore to limits access

    # method
    def student_details(self):
        return (f"{self.name} is in class {self.grade}, with percentage {self.percentage}%") # change any thing

# child class
# GraduateStudent is a child class which is inherited the Student class
class GraduateStudent(Student):
    def __init__(self,name, grade, percentage, stream): # i want to add stream
        # parameter from parent class
        super().__init__(name,grade,percentage) #if you dont want you can remove previous attribute
        # new parameter/ attribute use in child class
        self.stream=stream

    def student_details(self):  # method inherit, we can change the method name
        return (f"stream is {self.stream}")

# Object-GraduateStudent class
g_student1=GraduateStudent("Ram",12,96,"pcm")
# Object-Student class
stud1=Student("Arjun",11,92)

print(g_student1.student_details())
print(stud1.student_details())
```

**The name of the method is same here in both of the classes but Which class object we are making, We invoke that method here.**

**This is also a very vital use property of oops.**
