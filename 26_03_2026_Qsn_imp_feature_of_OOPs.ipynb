# **1: Library Book Management**

```
Description: Design a Book class to represent a library book. Each book has a title, author, ISBN, and availability
status (checked out or available). Implement methods to check out the book (if available), return the book, and
display its details.
```


```

```




````python
class Book:
    def __init__(self, title, author, ISBN):
        self.title=title
        self.author=author
        self.ISBN=ISBN
        self.is_available=True

    def check_out(self):
        if self.is_available:
            self.is_available=False
            return (f"{self.title} has been checked out.")
        else:
            return (f"{self.title} is already checked out")

    def return_book(self):
        #"The Great Gatsby" has been returned.
        self.is_available=True
        return (f"{self.title} has been returned.")

    def display_info(self):
        if self.is_available:
            print(f"Title: {self.title}\nAuthor: {self.author}\nISBN: {self.ISBN}\nStatus: Available")
        else:
            print(f"Title: {self.title}\nAuthor: {self.author}\nISBN: {self.ISBN}\nStatus: Not Available")
````

```
Description: Create a Student class that stores a student’s name, ID, and a list of their grades. Include methods
to add a new grade, calculate the average grade, and determine if the student is passing (average ¿= 60). The class
should also have a method to display a student’s report.
```





```
Tracking Student Grades
>>> student1 = Student("Alice", "S001")
>>> student1.add_grade(85)
>>> student1.add_grade(90)
>>> student1.add_grade(78)
>>> student1.get_average()
84.33
>>> student1.is_passing()
True
>>> student1.display_report()
Student: Alice (ID: S001)
Grades: [85, 90, 78]
Average: 84.33
Status: Passing
```



````python
class Student:
    def __init__(self,name,ID):
        self.name=name
        self.ID=ID
        self.grade=[]

    def add_grade(self,other):
        self.grade.append(other)

    def get_average(self):
        if len(self.grade)==0:
            return ("first enter the student marks details.")
        else:
            num=sum(self.grade)
            den=len(self.grade)
            avg=round(num/den,2)
            return avg

    def display_report(self):
        if len(self.grade)!=0:
            if self.get_average()>=60:
                return (f"Student: {self.name} (ID: {self.ID})\nGrades: {self.grade}\nAverage: {self.get_average()}\nStatus: Passing")
            else:
                return (f"Student: {self.name}\nGrades: {self.grade}\nAverage: {self.get_average()}\nStatus: Fail")
        else:
            return (f"Student: {self.name} (ID: {self.ID})\nGrades: {self.grade}\nAverage: Null\nStatus: Not determined")
````

## **Type 1: Method Overriding**

```
Child class modifies parent class method.
```



````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")
````

````python
class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)    # Call parent constructor
        self.student_id = student_id

    def display(self):
        super().display()               # Call parent method
        print(f"Student ID: {self.student_id}")

student = Student("Arjun", 20, "S123")
student.display()
````

````python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):           # Overriding
        print("Dog barks")

class Cat(Animal):
    def sound(self):           # Overriding
        print("Cat meows")

animals = [Animal(), Dog(), Cat()]
for animal in animals:
    animal.sound()
````

````python
class Multiply:
    def multiply(self,*args):
        mult=1
        for i in args:
            mult*=i

        return mult
````

````python
m2=Multiply()
m2.multiply(1,2,3,4)
````

````python
class Calculator:
    # Using default arguments
    def add(self, a, b=0, c=0):
        return a + b + c

    # Using *args (best way)
    def multiply(self, *numbers):
        result = 1
        for num in numbers:
            result *= num
        return result
````

````python
calc = Calculator()
print(calc.add(5))          # Output: 5
print(calc.add(5, 10))      # Output: 15
print(calc.add(5, 10, 15))  # Output: 30
````

````python
print(calc.multiply(2, 3))      # Output: 6
print(calc.multiply(2, 3, 4))   # Output: 24
````

### **Type 3: Operator Overloading**

```
Changing behavior of operators for user-defined classes.
or
Changing behavior of operators (+, −, ∗, etc.)
```



````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Overloading + operator
    def __add__(self, other):
        return self.marks + other.marks

s1 = Student("Arjun", 85)
s2 = Student("Rahul", 70)

print(s1 + s2)        # Output: 155
````

````python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Overloading + operator
    def __add__(self, other):
        return self.marks + other.marks

    # Overloading > operator
    def __gt__(self, other):
        return self.marks > other.marks

s1 = Student("Arjun", 85)
s2 = Student("Rahul", 70)

print(s1 + s2)        # Output: 155
print(s1 > s2)        # Output: True
````

# **What is Abstraction?**

```
Abstraction = Hiding implementation details, showing only essential features.

How to Implement Abstraction
We use:
• ABC (Abstract Base Class)
• @abstractmethod decorator
• From module: from abc import ABC, abstractmethod
```



````python
from abc import ABC, abstractmethod

class Shape(ABC):                    # Abstract class
    @abstractmethod
    def area(self):                  # Abstract method
        pass                         # No implementation
````

````python
class Circle(Shape):
    def area(self):                  # Must implement
        print("Area of circle")

class Rectangle(Shape):
    def area(self):                  # Must implement
        print("Area of rectangle")

# Using the classes
circle = Circle()
rectangle = Rectangle()
circle.area()        # Output: Area of circle
rectangle.area()     # Output: Area of rectangle
````

````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    @abstractmethod
    def start(self):
        """Every vehicle must have a start method"""
        pass

    @abstractmethod
    def stop(self):
        """Every vehicle must have a stop method"""
        pass

    @abstractmethod
    def fuel_type(self):
        """Every vehicle must specify fuel type"""
        pass

    # Concrete method (not abstract)
    def display_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}")
        print(f"Fuel Type: {self.fuel_type()}")
````

````python
ob1=Vehicle()
````

````python
class Car(Vehicle):
    def start(self):
        print(f"{self.brand} {self.model}: Starting engine... Vroom!")

    def stop(self):
        print(f"{self.brand} {self.model}: Stopping engine...")
````

````python
car = Car("Toyota", "Camry")
````

````python
class Car(Vehicle):
    def start(self):
        print(f"{self.brand} {self.model}: Starting engine... Vroom!")

    def stop(self):
        print(f"{self.brand} {self.model}: Stopping engine...")

    def fuel_type(self):
        return "Petrol/Diesel"
````

````python
# Using the classes
car = Car("Toyota", "Camry")
car.display_info()
car.start()
car.stop()
````

````python
class ElectricCar(Vehicle):
    def start(self):
        print(f"{self.brand} {self.model}: Starting electric motor... Silent!")

    def stop(self):
        print(f"{self.brand} {self.model}: Shutting down...")

    def fuel_type(self):
        return "Electricity"

    # Additional method specific to electric cars
    def charge_battery(self):
        print(f"{self.brand} {self.model}: Charging battery...")
````

````python
electric = ElectricCar("Tesla", "Model 3")
electric.display_info()
electric.start()
electric.charge_battery()
electric.stop()
````

## **What are Static Methods?**

```
Static methods = Methods that belong to class, not object.
```



**Make connection with class without self.**

````python
class Math:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def is_even(num):
        return num % 2 == 0

# Call without creating object
print(Math.add(5, 3))       # Output: 8
print(Math.is_even(4))      # Output: True
print(Math.is_even(5))      # Output: False
````

````python
ob1=Math()
# Also can be call with creating object
print(ob1.add(5, 3))       # Output: 8
print(ob1.is_even(4))      # Output: True
print(ob1.is_even(5))      # Output: False
````

````python
class Student:
    school_name = "IIT Palakkad"    # Static variable

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"{self.name}: {self.marks} - {Student.school_name}")

    @staticmethod
    def is_pass(marks):
        return marks >= 40
````

````python
# Using the class
s1 = Student("Arjun", 85)
s2 = Student("Rahul", 35)

s1.display()    # Output: Arjun: 85 - IIT Palakkad
s2.display()    # Output: Rahul: 35 - IIT Palakkad
````

````python
print(Student.is_pass(85))    # Output: True
print(Student.is_pass(35))    # Output: False
````
