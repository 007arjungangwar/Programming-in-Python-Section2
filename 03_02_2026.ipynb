## **Nested Conditional Statement**

When an if statement is written inside another if or else block.

üîπ Syntax

# **Find the sum of digit of given 3 digit number**

````python
Num=546
dig1=Num%10 # 6
Num//=10    # 54
dig2=Num%10 # 4
Num//=10    # 5
dig3=Num%10 # 5
sum=dig1+dig2+dig3
print(sum)
````

````python
# Get a 3 - digit number from user
number = int( input (" Enter a 3 - digit number : ") )
# Extract digits
digit1 = number % 10 # Units digit
number //= 10
digit2 = number % 10 # Tens digit
number //= 10
digit3 = number % 10 # Hundreds digit
# Calculate sum
digit_sum = digit1 + digit2 + digit3
print (f"Sum of digits : { digit1 } + { digit2 } + { digit3 } = { digit_sum }")
````

# **Find the sum of digit of given number(Note: we dont know the length of number 3,4,5....digits**

````python
num=int(input("Enter the positive number : "))
sum=0
while num>0:
  last_digit=num%10
  sum+=last_digit
  num//=10
print(sum)
````
