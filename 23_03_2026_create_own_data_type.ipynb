## **Create new data type with the name of the Fraction**

````python
class Fraction:
    def __init__(self,n,d):
        self.num=n
        self.den=d

    def __str__(self):
        return f"{self.num}/{self.den}"
````

### **Add some more functionality in the above fraction data type**

````python
class Fraction:
    def __init__(self,n,d):
        self.num=n
        self.den=d

    def __str__(self):            # print()
        return f"{self.num}/{self.den}"

    def __add__(self,other):      # +
        temp_num=self.num*other.den+other.num*self.den
        temp_den=self.den*other.den
        return f"{temp_num}/{temp_den}"

    def __sub__(self,other):      # -
        temp_num=self.num*other.den-other.num*self.den
        temp_den=self.den*other.den
        return f"{temp_num}/{temp_den}"

    def __mul__(self,other):      # *
        temp_num=self.num*other.num
        temp_den=self.den*other.den
        return f"{temp_num}/{temp_den}"

    def __truediv__(self,other):  # /
        temp_num=self.num*other.den
        temp_den=self.den*other.num
        return f"{temp_num}/{temp_den}"

    def __pow__(self,n):          # **
        temp_num=(self.num)**n
        temp_den=(self.den)**n
        return f"{temp_num}/{temp_den}"
````

````python
f1=Fraction(7,2)
f2=Fraction(1,2)
f3=Fraction(1,3)
````

````python
print(f1.num)
print(f1.den)
````

````python
print(f1)
print(f2)
print(f3)
````

````python
print(f1+f2)
print(f1-f2)
print(f1*f2)
print(f1/f2)
````

````python
print(f1**3)
````

````python
f1 = Fraction(1, 2)
f2 = Fraction(3, 4)

print(f1 + f2)    # 10/8
print(f1 - f2)    # -2/8
print(f1 * f2)    # 3/8
print(f1 / f2)    # 4/6
print(f1 ** 2)    # 1/4
````

### **Add some more functionality in the above Point data type**

````python
class Point:
    def __init__(self,x,y):
        self.x_cod=x
        self.y_cod=y

    def __str__(self):
        return f"<{self.x_cod},{self.y_cod}>"

    def __add__(self,other):
        return f"<{self.x_cod+other.x_cod},{self.y_cod+other.y_cod}>"

    def __sub__(self,other):
        return f"<{self.x_cod-other.x_cod},{self.y_cod-other.y_cod}>"


    def equlidean_distance(self,other):
        return round(((self.x_cod-other.x_cod)**2 + (self.y_cod-other.y_cod)**2)**0.5,2)
````

````python
p1=Point(1,2)
p2=Point(4,5)
print(p1)
print(p2)
print(p1+p2)
print(p1-p2)
print(p2-p1)
# p1.equlidean_distance(Point(4,5))
print(p1.equlidean_distance(p2))

print(p2.equlidean_distance(p1))
````

# **Create new data type with the name of the Line**

### **Write OOP classes to handle the following scenarios:**

- A user can create and view 2D coordinates
- A user can find out the distance between 2 coordinates
- A user can find find the distance of a coordinate from origin
- A user can check if a point lies on a given line
- A user can find the distance between a given 2D point and a given line

````python
class Point:
    def __init__(self,x,y):
        self.x_cod=x
        self.y_cod=y

    def __str__(self):
        return f"<{self.x_cod},{self.y_cod}>"

    def __add__(self,other):
        return f"<{self.x_cod+other.x_cod},{self.y_cod+other.y_cod}>"

    def __sub__(self,other):
        return f"<{self.x_cod-other.x_cod},{self.y_cod-other.y_cod}>"


    def equlidean_distance(self,other):
        return round(((self.x_cod-other.x_cod)**2 + (self.y_cod-other.y_cod)**2)**0.5,2)


class Line:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def __str__(self):
        return f"{self.a}x+{self.b}y+{self.c}=0"

    def point_lies_on_line(self,other):
        if self.a*other.x_cod+self.b*other.y_cod+self.c ==0:
            return f"point <{other.x_cod},{other.y_cod}> lies on a line {self.a}x+{self.b}y+{self.c}=0."

        else:
            return f"point <{other.x_cod},{other.y_cod}> does`t lie on a line {self.a}x+{self.b}y+{self.c}=0."


    def shortest_distance_from_point(self,other):
        temp_num=abs(self.a*other.x_cod+self.b*other.y_cod+self.c)
        temp_den=(self.a**2 +self.b**2)**0.5
        value=round(temp_num/temp_den,2)
        return f"shortest distance is {value} from point <{other.x_cod},{other.y_cod}> to the line {self.a}x+{self.b}y+{self.c}=0."
````

````python
l1=Line(1,2,3)
print(l1)

p1=Point(1,1)
print(p1)

p2=Point(-1,-1)
print(p2)
````

````python
print(l1.point_lies_on_line(p1))
print(l1.point_lies_on_line(p2))
````

````python
print(l1.shortest_distance_from_point(p1))
print(l1.shortest_distance_from_point(p2))
````
